```python
# Load CSV
df = pd.read_csv("best_solution.csv")

DEPOT_LOCATION = 12

depot_index = list(df[df["Location_ID"] == DEPOT_LOCATION].index)
routes = []
for i in range(len(depot_index)-1):
    routes.append(df.iloc[depot_index[i]:depot_index[i+1]+1,:])
```

```python
for i in range(len(routes)):
    plt.plot(routes[i]["X"], routes[i]["Y"], marker='o', linestyle='-')
    plt.scatter(routes[i]["X"], routes[i]["Y"], color='red')  # Highlight points
    
    # Add labels for each point
    for idx, row in routes[i].iterrows():
        plt.text(row["X"], row["Y"], int(row["Location_ID"]), fontsize=10, ha='right', va='bottom')

plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Best Found TSP Solution Path")
plt.grid()
plt.show()
```
